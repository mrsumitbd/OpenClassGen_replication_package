```python
from os import listdir
from os.path import isfile, join
import pandas as pd
```

```python
zip_data_files = [f for f in listdir('./data/') if (isfile(join('./data/', f)) and f.endswith(f".gz"))]
```

```python
zip_data_files
```

```python
df_list = []
for file in zip_data_files:
    df = pd.read_json(f'data/{file}', lines=True, compression='gzip')
    df_list.append(df)

combined_df = pd.concat(df_list, ignore_index=True)
```
